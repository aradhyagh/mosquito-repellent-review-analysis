```python
class nlpreviews:
    def __init__(self):
        import re
        import spacy
        self.spacy = spacy
        self.nlp = spacy.load("en_core_web_sm")

    def aspect_detect(self, text):
        text = text.lower()
        doc = self.nlp(text)
        aspects = []
        for chunk in doc.noun_chunks:
            # Filter out stopwords and lemmatize each token in the chunk
            filtered_tokens = [
                token.lemma_ for token in chunk if not token.is_stop and token.is_alpha
            ]
            if filtered_tokens:  # Add only if there's something left
                aspects.append(' '.join(filtered_tokens))
        return ' '.join(aspects)
```

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U00002500-\U00002BEF"  # Chinese characters
        "\U00002702-\U000027B0"
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "\U0001f926-\U0001f937"
        "\U00010000-\U0010ffff"
        "\u2640-\u2642"
        "\u2600-\u2B55"
        "\u200d"
        "\u23cf"
        "\u23e9"
        "\u231a"
        "\ufe0f"  # Dingbats
        "\u3030"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
import spacy
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Load SpaCy and VADER
nlp = spacy.load("en_core_web_sm")
sid = SentimentIntensityAnalyzer()

def extract_aspect_sentiments(text):
    doc = nlp(text)
    aspect_sentiments = {}

    for sent in doc.sents:
        sentence = sent.text
        parsed_sent = nlp(sentence)
        for token in parsed_sent:
            if token.pos_ == "NOUN":
                # Get modifiers (adjective or adverb)
                modifiers = [child.text for child in token.children if child.dep_ in ("amod", "acomp", "advmod")]
                if modifiers:
                    phrase = f"{' '.join(modifiers)} {token.text}"
                else:
                    phrase = token.text
                sentiment_score = sid.polarity_scores(sentence)["compound"]
                if sentiment_score >= 0.05:
                    sentiment = "positive"
                elif sentiment_score <= -0.05:
                    sentiment = "negative"
                else:
                    sentiment = "neutral"
                aspect_sentiments[token.text.lower()] = sentiment

    return aspect_sentiments
```

```python
df["aspect_sentiments"]=df["Body"].apply(extract_aspect_sentiments)
```

```python
# Count Vectorizer
count_vect = CountVectorizer(stop_words='english')
dtm = count_vect.fit_transform(df['Body'])

# LDA Model
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(dtm)

# Topic Display Function
def display_topics(model, feature_names, n_top_words=5):
    for topic_idx, topic in enumerate(model.components_):
        print(f"LDA Topic {topic_idx+1}: " + ", ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

display_topics(lda, count_vect.get_feature_names_out())
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

model_name = "bhadresh-savani/bert-base-go-emotion"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def predict_emotions(text, top_k=3):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True)

    # Get model predictions
    with torch.no_grad():
        logits = model(**inputs).logits
        probs = torch.sigmoid(logits)[0]  # Apply sigmoid for multi-label classification

    # Get top k emotions
    top_probs, top_indices = torch.topk(probs, k=top_k)
    labels = [model.config.id2label[idx.item()] for idx in top_indices]
    scores = [prob.item() for prob in top_probs]

    # Return results as a list of tuples (label, score)
    return list(zip(labels, scores))
```

```python
def emotion(text):
  abc = predict_emotions(text)  # Use the function parameter 'text'
  return abc

df["emotions_review"] = df["Body"].apply(emotion)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon (only once)
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define function to get sentiment label
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return "neutral"  # default for empty or invalid entries
    score = sid.polarity_scores(text)["compound"]
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

# Apply to 'review' column
df['predicted_sentiment'] = df['Body'].apply(get_sentiment)
```
